```python
import sounddevice as sd
from scipy.io.wavfile import write, read
from scipy.fftpack import fft
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = 44100  # Sample rate
seconds = 5  # Duration of recording
# Channel L = index 0
# Channel R = index 1

myrecording = sd.rec(int(seconds * fs), samplerate=fs, channels=2)
sd.wait()  # Wait until recording is finished
write('output.wav', fs, myrecording)  # Save as WAV file 
```

```python
my_sound = myrecording[fs:,1]
plt.figure()
plt.plot(my_sound)
plt.show()
```

```python
# Count number of ticks in the recording,
# assuming that each tick lasts less than 1000 samples
# (Interval of 0.1 seconds (10 Hz) is 4410 samples)
def count_ticks(sound, threshold):
    count = 0
    i = 0
    while i < len(sound):
        sample = sound[i]
        if sample > threshold:
            count += 1
            i += 1000
        else:
            i += 1
    return count
```

```python
# Returns the index of the first peak
def find_first_peak(sound, threshold):
    peak_val = threshold
    max_i = None
    i = 0
    while i < len(sound):
        sample = sound[i]
        if sample > threshold:
            peak_val = max(sound[i:i+1000])
            return np.where(sound==peak_val)[0][0]
        else:
            i += 1
    return max_i

# Returns the index of the last peak
def find_last_peak(sound, threshold):
    i = len(sound) - 1
    max_i = None
    peak_val = threshold
    while i >= 0:
        sample = sound[i]
        if sample > threshold:
            peak_val = max(sound[i-1000:i])
            return np.where(sound==peak_val)[0][0]
        else:
            i -= 1
    return max_i
```

```python
def get_beat_error(freq, target_freq=6):
    num_seconds_a_day = 24*60*60
    one_clock_incment = 1/target_freq
    num_clock_seconds_a_day = freq * num_seconds_a_day * one_clock_incment
    return num_clock_seconds_a_day - num_seconds_a_day
```

```python
plt.figure()
plt.plot(my_sound[find_first_peak(my_sound, 0.2):find_last_peak(my_sound, 0.2)])
plt.show()
```

```python
trimmed = my_sound[find_first_peak(my_sound, 0.2):find_last_peak(my_sound, 0.2)]
count_ticks(trimmed, 0.2)
```

```python
watch_recording = trimmed
N = len(watch_recording)
T = 1/fs
time_stamp = np.linspace(0, N/fs, N)
yf = fft(watch_recording)
xf = np.linspace(0.0, 1.0/(2.0*T), N//2)
```

```python
plt.figure()
plt.plot(time_stamp, watch_recording)
plt.show()
```

```python
plt.figure()
plt.plot(xf, 2.0/N * np.abs(yf[0:N//2]))
plt.xlim(0, 20)
plt.show()
```

```python
lower_i = 0
upper_i = 0
i = 0
while xf[i] < 6.5:
    if xf[i] < 5.5 and xf[i+1] >= 5.5:
        lower_i = i
    i += 1
upper_i = i
```

```python
freq = xf[np.where(max((2.0/N * np.abs(yf[0:N//2]))[lower_i:upper_i]) == (2.0/N * np.abs(yf[0:N//2])))[0][0]]
freq
```

```python
get_beat_error(freq)
```
